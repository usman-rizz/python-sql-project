# <b>Loading dataset</b> 

```python
df = pd.read_csv('netflix_titles.csv', encoding="utf-8")
```

# <b>Connecting dataset to sql server</b>

```python
import sqlalchemy as sal

engine = sal.create_engine(
    "mssql+pyodbc://@Usman/netflix--database"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
)

conn = engine.connect()
print("Connected successfully")
```

# <b>Loading the dataset into database</b>

```python
# Write to raw table
df.to_sql('netflix_raw', con=conn, index=False, if_exists='replace')
conn.close()

# not re-run this cell again buz file is already loaded to the database thats why I commented out
```
